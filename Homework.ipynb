## Plotting Data Visualisasi Histogram

```python
fig = plt.figure()
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16,10))

axes[0].hist(tugas1['AAPL'])
axes[0].set_title('AAPL')

axes[1].hist(tugas1['IBM'])
axes[1].set_title('IBM')

axes[2].hist(tugas1['CSCO'])
axes[2].set_title('CSCO')

axes[3].hist(tugas1['MSFT'])
axes[3].set_title('MSFT')

for ax in axes.flat:
    ax.set(xlabel='Stock', ylabel='Jumlah')

#menyembunyikan label x dan y yang saling betabrakan antar plot
for ax in axes.flat:
    ax.label_outer()
```

## Plotting Data Visualisasi Scatter

```python
fig, axes = plt.subplots(figsize=(16,10))

axes.scatter(x=tugas1['IBM'], y=tugas1['CSCO'], color='r')
axes.scatter(x=tugas1['IBM'], y=tugas1['AAPL'], color='g')
axes.scatter(x=tugas1['IBM'], y=tugas1['MSFT'], color='y')
axes.scatter(x=tugas1['CSCO'], y=tugas1['AAPL'], color='b')
axes.scatter(x=tugas1['CSCO'], y=tugas1['MSFT'], color='brown')
axes.set(title='Scatter Plot')
axes.set(xlabel='Stock')
axes.set(ylabel='Jumlah')
```
